```python
#Importar Librerias
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
#Renombrar el nombre de las columnas
df = df.rename({'survived': 'Sobreviviente','pclass':'ClasePasaje',
                    'sex':'Sexo','age':'Edad',
                    'sibsp': 'HnsEsposo', 'parch':'PadresHijos',
                    'fare': 'Tarifa','embarked':'Embarcacion',
                    'class': 'Clase','who':'Genero','adult_male':'HombreAdulto',
                    'deck': 'Cubierta', 'embark_town': 'LugarEmbarcacion',
                    'alive': 'Vivo','alone':'Solo'}, axis=1)
df
```

```python
df.dtypes
```

## ***I. Método No Gráfico***

## **Relación entre dos variables**

### ***Tabla Cruzada***

```python
#Creamos una tabla cruzada entre el clase y cubierta
pd.crosstab(df['Clase'], df['### Completar'])  ### Completar
```

```python
#Creamos una tabla cruzada entre el clase y cubierta
#agregar margins: devuelve la suma (el tipo depende de las cols)
pd.crosstab(df['### Completar'], df['### Completar'], margins=True)
```

```python
#Para normalizar a % se usa normalize= all
pd.crosstab(df['### Completar'], df['### Completar'], normalize='all', margins=True)
```

## **Relación entre variables cuantitativas**

### ***Covarianza***

*   La función `cov()` calcula la covarianza entre todas las variables cuantitativas

```python
# Seleccionar solo las columnas numéricas
numeric_df = df.select_dtypes(include=['number'])

# Calcular la matriz de covarianza
cov_matrix = numeric_df.cov()

# Mostrar la matriz de covarianza
cov_matrix
```

### ***Correlación***

*   La función `corr()` calcula la correlación entre todas las variables cuantitativas

```python
numeric_df.corr()
```
