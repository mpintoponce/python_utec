```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
excel_path='/content/CUMPLIMIENTO PLAN.xlsm'
```

```python
df_clean = pd.read_excel(excel_path, sheet_name='IW38 PLAN SEMANAL')
```

```python
df_clean
```

```python
# Limpia espacios y cambia a mayúscula la primera letra
df_clean['ESTADO OTS'] = df_clean['ESTADO OTS'].str.strip().str.capitalize()
```

```python
import matplotlib.pyplot as plt

colores_dict = {
    'Abierto': '#FF6666',        # rojo
    'Cerrado': '#6699FF',        # azul
    'Reprogramado': '#FFEB66'    # amarillo
}


# Cuenta los estados después de limpiar
status_counts = df_clean['ESTADO OTS'].value_counts()

# Genera colores en orden correcto
colores = [colores_dict.get(estado, '#cccccc') for estado in status_counts.index]

# Graficar
status_counts.plot(kind='pie', colors=colores, autopct='%1.1f%%', title='Estado de Ots')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt



# Colores personalizados (puedes modificar estos valores)

colores = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']



# Gráfico con colores personalizados

status_counts = df_clean['Status de sistema'].value_counts()

status_counts.plot(

    kind='pie',

    autopct='%1.1f%%',

    title='Distribución por Status de Sistema',

    colors=colores

)

plt.ylabel('')

plt.show()
```

```python
clase_counts = df_clean['Clase de orden'].value_counts()
clase_counts.plot(kind='bar', title='Distribución de Órdenes por Clase')
plt.ylabel('Número de Órdenes')
plt.xlabel('Clase de Orden')
plt.xticks(rotation=45)
plt.show()
```

```python
prioridad_counts = df_clean['Prioridad'].value_counts().sort_index()
prioridad_counts.plot(kind='bar', title='Distribución de Órdenes por Prioridad')
plt.ylabel('Número de Órdenes')
plt.xlabel('Prioridad (1=alta, 3=baja)')
plt.show()
```

```python
status_counts = df_clean['Status de sistema'].value_counts()
status_counts.plot(kind='pie', autopct='%1.1f%%', title='Distribución por Status de Sistema')
plt.ylabel('')
plt.show()
```

```python
df_clean.groupby('Prioridad')['Cantidad'].mean().plot(kind='bar', title='Cantidad Promedio de Trabajo por Prioridad')
plt.ylabel('Horas promedio de trabajo')
plt.xlabel('Prioridad')
plt.show()
```

```python
top_puestos = df_clean['Puesto de trabajo'].value_counts().head(10)
top_puestos.plot(kind='barh', title='Top 10 Puestos de Trabajo más Comunes')
plt.xlabel('Número de Órdenes')
plt.ylabel('Puesto de Trabajo')
plt.show()
```

```python
df_clean['Holgura total'].plot(kind='hist', bins=20, title='Distribución de Holgura Total')
plt.xlabel('Horas de holgura')
plt.ylabel('Frecuencia')
plt.show()
```

```python
df_clean['Texto breve operación'].str[:10].value_counts().head(10).plot(kind='bar', title='Operaciones más Comunes (primeros 10 caracteres)')
plt.ylabel('Número de Órdenes')
plt.xticks(rotation=45)
plt.show()
```
