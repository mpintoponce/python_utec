```python
import pandas as pd
import seaborn as sns
```

## **Importar Datos**

```python
df = pd.read_csv('https://data.insideairbnb.com/greece/attica/athens/2025-03-18/data/listings.csv.gz')
```

```python
df2 = df[["id","host_id", "listing_url","host_name", "host_is_superhost", "name", "neighbourhood", "neighbourhood_cleansed", "host_response_rate", "latitude", "longitude", "room_type", "price", "number_of_reviews", "review_scores_rating" ,"accommodates","bathrooms" ]].copy()
```

```python
df2.info()
```

```python
df2.head()
```

```python
df2.dropna(inplace=True)
df2.info()
```

```python
#Limpiamos
def limpiar(a):
  return str(a).replace("$","").replace(",","")

df2["price"] = df2["price"].apply(limpiar).astype(float)
df2["host_response_rate"] = df2["host_response_rate"].str.replace("%","").astype(float)

df2.head()
```

```python
df2.describe()
```

```python
df_filter2= df_filter[(df_filter['review_scores_rating']>=4.8) & (df_filter['number_of_reviews']>=200)]
df_filter2.info()
```

```python
df_c = pd.read_csv('https://data.insideairbnb.com/greece/attica/athens/2025-03-18/data/calendar.csv.gz')
df_c
```

```python
df_c["date"] = pd.to_datetime(df_c["date"])
df_c.info()
```

```python
df_c = df_c[
    (df_c['available'] == 't') &
    (df_c['date'] >= pd.to_datetime("2025-09-15")) &
    (df_c['date'] <= pd.to_datetime("2025-09-30"))
]

df_c
```

```python
df_c_count = df_c.groupby('listing_id').count()
df_c_count = df_c_count[df_c_count['date'] == 16]
df_c_count[]
```

```python
at_2025 = pd.merge(df_filter2, df_c_count, left_on='id', right_on='listing_id',how='left')
at_2025
```

```python
at_2025= at_2025[(at_2025['date']==16)]
at_2025
```

```python
at_2025.info()
```
