```python
# Datos de un cliente
ingreso_mensual = 4500 # en d√≥lares
anios_cliente = 7
productos = 4  # tarjetas, cuentas, seguros, etc.
operaciones_fallidas = 3
edad = 45

# C√°lculo de puntaje de valor usando operadores aritm√©ticos

# 1. Ingreso anual
ingreso_anual = ingreso_mensual * 12

# 2. Bonificaci√≥n por fidelidad (a√±os de cliente al cuadrado)
bono_fidelidad = anios_cliente ** 2

# 3. Penalidad por operaciones fallidas (cada una resta 100 puntos)
penalidad = operaciones_fallidas * 100

# 4. Puntaje base (ingreso anual dividido entre edad)
puntaje_base = ingreso_anual / edad

# 5. Puntaje total antes de normalizar
puntaje_total = puntaje_base + bono_fidelidad - penalidad + (productos * 250)

# 6. Normalizamos con divisi√≥n entera entre 100
segmento = puntaje_total // 100

# 7. Usamos m√≥dulo para ver si el puntaje es redondo
es_redondo = puntaje_total % 100 == 0

# Resultados
print("Puntaje total:", round(puntaje_total, 2))
print("Segmento (0 = bajo valor, 1 = medio, 2 o m√°s = alto valor):", int(segmento))
print("¬øEl puntaje es redondo?", es_redondo)
```

Ejemplo

Comparar la tasa ofrecida al cliente con la tasa est√°ndar del banco y calcular el spread (diferencia), redondeando el resultado y mostrando su valor absoluto si es negativo.

```python
# Datos base
standar_rate = 0.105  # 10.5%
client_rate = float(input("Ingrese la tasa ofrecida al cliente (por ejemplo 0.098 para 9.8%): "))

# C√°lculo del spread
spread = client_rate - standar_rate
spread_rounded = round(spread, 4)  # redondear a 4 decimales
spread_absolute = abs(spread_rounded)  # valor absoluto del spread

# Mostrar resultado
print("\nTasa est√°ndar del banco:", round(standar_rate * 100, 2), "%")
print("Tasa ofrecida al cliente:", round(client_rate * 100, 2), "%")
print("Diferencia de tasas (spread):", spread_rounded)
print("Valor absoluto del spread:", spread_absolute)
```

```python
# Datos base
standar_rate = 0.105  # 10.5%
client_rate = float(input("Ingrese la tasa ofrecida al cliente (por ejemplo 0.098 para 9.8%): "))

# C√°lculo del spread
spread = client_rate - standar_rate
spread_rounded = round(spread, 4)  # redondear a 4 decimales
spread_absolute = abs(spread_rounded)  # valor absoluto del spread

# Mostrar resultado
print("\nTasa est√°ndar del banco:", round(standar_rate * 100, 2), "%") # Se coloca '\n' cuando quiero que el texto agregue un espacio encima de la l√≠nea
print("Tasa ofrecida al cliente:", round(client_rate * 100, 2), "%")
print("Diferencia de tasas (spread):", spread_rounded)
print("Valor absoluto del spread:", spread_absolute)
```

**¬øQu√© hacen round() y abs() aqu√≠?**


* round(valor, 4) ‚Üí redondea el resultado a 4 decimales, √∫til para porcentajes y precisi√≥n en precios.
* abs(valor) ‚Üí devuelve el valor positivo de una diferencia, √∫til para analizar la magnitud del spread sin importar el signo.

üß† **Dato curioso:**

En el mundo bancario, el spread es una de las principales fuentes de ingresos. Aunque parezca peque√±o (como 0.0025 o 0.25%), aplicado a millones de d√≥lares genera grandes m√°rgenes. ¬°Por eso la precisi√≥n en el c√°lculo es clave!

# ¬øQu√© es una lista en Python?

Una lista es una estructura de datos que permite guardar varios valores dentro de una sola variable.
Es como una fila de casilleros numerados: cada casillero tiene un valor, y puedes acceder, cambiar o eliminar valores f√°cilmente.

**¬øPara qu√© sirve una lista?**

* Para guardar muchos datos relacionados (como montos de pr√©stamos, edades, tasas de inter√©s, etc.)
* Para analizar informaci√≥n en bloque, como calcular promedios, contar elementos o buscar tendencias.
* Para almacenar resultados intermedios o hist√≥ricos en an√°lisis de datos.

Ejemplo pr√°ctico

```python
ingreso = 2800
edad = 30
historial_morosidad = 1

if ingreso >= 2000 and historial_morosidad == 0:
    print("Recomendado: Cr√©dito personal")
elif ingreso >= 1000 and historial_morosidad <= 1:
    print("Recomendado: Cr√©dito con garant√≠a")
else:
    print("Recomendado: No apto por ahora")
```

 **¬øCu√°l es la diferencia entre elif y else en Python?**

1. elif (significa ‚Äúelse if‚Äù ‚Üí ‚Äúsi no, pero si‚Ä¶‚Äù):

* Eval√∫a una condici√≥n espec√≠fica, solo si la anterior fue falsa.
* Puedes tener varios elif seguidos.
* Cada elif tiene su propia condici√≥n.

2. else (significa ‚Äúsi no‚Ä¶‚Äù a secas):
* Es una opci√≥n por defecto, cuando ninguna de las condiciones anteriores se cumple.
* No tiene condici√≥n, es como un "plan B".
